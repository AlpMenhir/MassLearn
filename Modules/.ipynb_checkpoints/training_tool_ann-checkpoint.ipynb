```python
import os
import re

os.chdir(r'C:\Users\Ronam\Desktop\Working_folder\Bioinformatic_projects\Molecular Network project\Masslearn')
# TODO change this to masslearn directory in fmanager

from ipywidgets import widgets
import Modules.statplot as sp
import Modules.cleaning as cl
import matplotlib.pyplot as plt
from ipywidgets import interact_manual
from scipy.stats import zscore
import pandas as pd
import numpy as np
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean
from scipy.optimize import curve_fit
from scipy.stats import norm, pearsonr
from tqdm import tqdm
import time
import pickle

import plotly.graph_objs as go
from plotly.subplots import make_subplots


"""
VERY IMPORTANT: based on the rt_shift and how you train your data, you must adapt all features from a csv file after mzmine
with +- 0.05, by the way this info should be in the model's name

"""

feat = pd.read_csv(r'C:\Users\Ronam\Desktop\Working_folder\Bioinformatic_projects\Data\RNN_training_set\training.csv')
mzML_folder = r'C:\Users\Ronam\Desktop\Working_folder\Bioinformatic_projects\Data\RNN_training_set\mzML'

feat = pd.read_csv(r'C:\Users\Ronam\Desktop\Working_folder\Bioinformatic_projects\Molecular Network project\feature\fl_ms1.csv')
mzML_folder = r'C:\Users\Ronam\Desktop\Working_folder\Bioinformatic_projects\Molecular Network project\mzML'



class TrainingToolRNN():
    """
    The training tool ann is used for discriminating feature shape quality using RNN
    """
    def __init__(self, Datafile, mzML_folder):       
        self.datafile = Datafile # data is a df with ms1 features, the csv generated by mzmine3
        self.data = {'sample':[], 'rt_range_min':[], 'rt_range_max': [], 'mz_range_min':[], 'mz_range_max': []}
        last_sample = ''
        for colname in self.datafile.columns:  
            fill_col = 0 # this varaible is to check if the 4 parameters (rt min max and mz min max) are filled for each feature from a sample
            if 'datafile' in colname:  
                sample = colname.split(':')[1]
                if last_sample != sample:                    
                    last_sample = sample
                    mz_min = self.datafile.loc[:,f'datafile:{sample}:mz_range:min'].dropna().to_list()
                    mz_max = self.datafile.loc[:,f'datafile:{sample}:mz_range:max'].dropna().to_list()
                    rt_min = self.datafile.loc[:,f'datafile:{sample}:rt_range:min'].dropna().to_list()
                    rt_max = self.datafile.loc[:,f'datafile:{sample}:rt_range:max'].dropna().to_list()
                    for i in range(len(mz_min)): # add in sample column
                        self.data['sample'].append(sample)
                    self.data['rt_range_min'].extend(rt_min)
                    self.data['rt_range_max'].extend(rt_max)
                    self.data['mz_range_min'].extend(mz_min)
                    self.data['mz_range_max'].extend(mz_max)
        self.data = pd.DataFrame(self.data)
        self.data = self.data.sample(frac=1, random_state=42) # randomize the df rows, 42 is a seed
        self.labels = [] # The output taking labels information
        self.spectra = {}
        self.arrays = None
        self.features = [] # The output taking matching information (m/z diff, rt diff, similarity metric, euclidean distance)
        self.block = False
        self.current_index = 0
        

        # Create a progress bar using tqdm
        progress_bar = tqdm(total=len(self.data['sample'].unique()))
        for c, sample in enumerate(self.data['sample'].unique()): #get samples names to take spectra from mzml files
            spec = cl.Spectra(os.path.join(mzML_folder ,sample))
            spec.extract_peaks()
            self.spectra[sample] = spec
            progress_bar.update(1) # Update the progress bar
        progress_bar.close()
        self.display_next_feature()
        
    def plot_features_plotly(self, Spectra, Diff, Length, Int_max):
        spectra = Spectra
        diff = Diff

        # Divide into three parts
        left = Spectra[:Diff[0]]
        feature = Spectra[Diff[0]:Diff[1]]
        right = Spectra[Diff[1]:]

        # Create a constrained version of the 'feature' values between 0 and 1
        relative_feature = (feature/max(feature))*Int_max

        # Create a new x-axis array
        x = np.concatenate([
            np.linspace(0, 9/20, len(left), endpoint=False), 
            np.linspace(9/20, 11/20, len(feature), endpoint=False),
            np.linspace(11/20, 1, len(right)) 
        ])

        # Plot data
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=x, y=spectra, mode='lines', name='Normalized Intensities', line=dict(width=0.75)))
        fig.add_trace(go.Scatter(x=x[Diff[0]:Diff[1]], y=relative_feature, mode='lines', name='Relative Feature', line=dict(width=1)))
        
        # Add vertical lines
        fig.add_shape(type="line", x0=9/20, y0=0, x1=9/20, y1=1, line=dict(color="Black", width=1))
        fig.add_shape(type="line", x0=11/20, y0=0, x1=11/20, y1=1, line=dict(color="Black", width=1))

        # Set x-axis and y-axis properties
        fig.update_layout(
        width=1000,  # Width in pixels
        height=400,  # Height in pixels
        xaxis=dict(range=[0, 1], showgrid=True), 
        yaxis=dict(range=[0, Int_max], showgrid=True, title='Intensity'),
        title=f'Feature nb of scans: {Length}',
        xaxis_title='rt axis',
        yaxis_title='Normalized Intensity',
        legend=dict(y=-0.2),  # Place the legend below the graph
        legend_orientation="h",  # Make the legend horizontal
        )

        # Show the figure
        fig.show()
        
        x_skewness = skew(feature)
        x_kurtosis = kurtosis(feature)

        print('Skewness:', x_skewness)
        print('Kurtosis:', x_kurtosis)
        

        # Assume your time or X-axis values
        time = np.linspace(0, 1, len(feature))
        
        # Fit a degree 3 polynomial to your signal
        coefficients = np.polyfit(time, feature, 3)

        # Generate the baseline
        baseline = np.polyval(coefficients, time)

        # Subtract the baseline
        signal_without_baseline = feature - baseline

        from scipy.signal import savgol_filter
        smoothed_signal = savgol_filter(feature, 5, 2)
        
        # Plot the original signal and the baseline
        plt.figure(figsize=(10,6))
        plt.plot(time, feature, label='Original')
        plt.plot(time, smoothed_signal, label='Smoothed_signal')
        plt.plot(time, baseline, label='Baseline')

        plt.legend()
        plt.show()


    # Function to fill 0 intensities values with border intensities values
    def fix_zeros(self, data):
        for i in range(len(data)):
            if data[i] == 0:
                if i == 0:  # If it's the first element
                    next_val = next((x for x in data[i+1:] if x != 0), 0)
                    data[i] = next_val
                elif i == len(data) - 1:  # If it's the last element
                    prev_val = next((x for x in reversed(data[:i]) if x != 0), 0)
                    data[i] = prev_val
                else:  # If it's any other element
                    prev_val = next((x for x in reversed(data[:i]) if x != 0), 0)
                    next_val = next((x for x in data[i+1:] if x != 0), 0)
                    data[i] = (prev_val + next_val) / 2
        return data

    # Function to obtain intensities for plotting features
    def spectra_get_ms_intensities(self, feature, RT_shift = 3):
        rt_shift = RT_shift # this is to increase the observation window of the features
        sample = feature['sample']
        rt_min = float(feature['rt_range_min'])
        rt_max = float(feature['rt_range_max'])
        mz_tolerance = 0.002 # mz tolerance is to take all mz around the feature mz values with a tolerance
        mz_min = float(feature['mz_range_min'])
        mz_max = float(feature['mz_range_max'])
        spec = self.spectra[sample]       
        scans = [] # scans are minimal and maximal scans index based of the rt range INCLUDING rt shift
        for rt_value  in [rt_min - rt_shift, rt_max + rt_shift]:
            match = 100 # match is 100min value to begin with a very high impossible rt value
            for idx, rt in enumerate(spec.rt1): # the idea is to have the lowest rt difference find the corresponding rt value of a scan
                difference = float(rt_value) - float(rt)  # take the difference of rt
                if abs(difference) < abs(match):
                    match = difference
                else: # if the difference increases rather than decreasing we have found the right rt, because the loop is in increase rt duration order
                    scans.append(idx-1)
                    break
        feature_scans = [] # scans are minimal and maximal scans index based of the rt range WITHOUT rt shift, so the min and max deduced from mzmine
        for rt_value  in [rt_min, rt_max]:
            match = 100 # match is 100min value to begin with a very high impossible rt value
            for idx, rt in enumerate(spec.rt1): # the idea is to have the lowest rt difference find the corresponding rt value of a scan
                difference = float(rt_value) - float(rt)  # take the difference of rt
                if abs(difference) < abs(match):
                    match = difference
                else: # if the difference increases rather than decreasing we have found the right rt, because the loop is in increase rt duration order
                    feature_scans.append(idx-1)
                    break
        if len(scans) == 2: 
            diff = [feature_scans[0] - scans[0] - 1]# take the scan difference with rt_shift, it will be for the plot to display window bars
            diff.append( feature_scans[1] - scans[0] + 1) # diff contain the scan difference from the min and max rt
            
            intensities = []
            
            scan_data = spec.peakarray1[scans[0]:scans[1]]

            # Convert the scan_data to numpy array
            scan_data = np.array(scan_data)

            # Get mz values and intensities
            mz_values = scan_data[:,:,0]
            intensities = scan_data[:,:,1]

            # Create masks for lower and upper bounds
            lower_mask = mz_values >= (mz_min - mz_tolerance)
            upper_mask = mz_values <= (mz_max + mz_tolerance)

            # Apply both masks to get the matching mz values and corresponding intensities
            matches_mz = np.where(lower_mask & upper_mask, mz_values, np.nan)
            matches_intensities = np.where(lower_mask & upper_mask, intensities, 0)

            # Get the maximum intensities for each scan (replace nan with 0)
            max_intensities = np.nanmax(matches_intensities, axis=1)
            max_intensities[np.isnan(max_intensities)] = 0

            return max_intensities.tolist(), diff
        else:
            return [], []
    
    
    # Function to handle the button click
    def on_button_click(self, label):
        if self.current_index == len(self.data):
            if self.block == False:
                self.labels.append(label)  # Map different labels 'high, low, noise, double high, ...
            print("All features have been labeled.")
            self.block = True # prevent for adding new labels
        else:
            if self.block == False:
                left, feature, right = self.arrays
                self.labels.append(label)  # Map 'Yes' to 1, 'No' to 0, and 'Skip' to None
                self.features.append([left, feature, right, len(feature), feature.max()])
                self.display_next_feature()
        if self.current_index % 10 == 0: # every 10 features, the trained model is saved
            with open("Cache/model_trained_features.pickle", "wb") as file:
                pickle.dump(self.features, file)
            with open("Cache/model_trained_labels.pickle", "wb") as file:
                pickle.dump(self.labels, file)
        
        print(f'{self.current_index + 1}/{len(self.data)}')        
        print(f'Length labels, feature = {len(self.labels)}, {len(self.features)}')
        print(f'Noise: {self.labels.count("Noise")}\nFeature: {self.labels.count("Feature")}')
        
    
    # Function to display the next feature
    def display_next_feature(self):
        feature = []
        while len(feature) < 8:  
            feature = self.data.iloc[self.current_index,]
            self.current_index += 1
            intensities, diff = self.spectra_get_ms_intensities(feature)
            if intensities:
                feature = intensities[diff[0]:diff[1]] # intensities values for the feature
            if (len(feature) >= 8) :     
                intensities = np.array(intensities)
                feature = intensities[diff[0]:diff[1]]
                left = intensities[:diff[0]] # intensities values at the left of the min rt, idea is to add this information to the decision network
                right = intensities[diff[1]:] # intensities values at the right of the min rt
                self.plot_features_plotly(intensities, diff, len(feature), intensities.max())
                self.arrays = left, feature, right
                break
        
        
tool = TrainingToolRNN(feat, mzML_folder)
J = interact_manual(tool.on_button_click, label=['Feature - Feature', 'Feature - Noise', 'Noise - Feature', 'Noise - Noise'])
```

```python
from scipy.interpolate import interp1d

def resample_curve(curve, num_points):
    if np.isnan(curve).all():
        return np.zeros(num_points)
    # Create a function that linearly interpolates between the points in the curve
    interpolation = interp1d(range(len(curve)), curve)

    # Use the function to interpolate at num_points evenly spaced intervals between the start and end of the curve
    new_curve = interpolation(np.linspace(0, len(curve) - 1, num_points))

    return new_curve

num_points = 100  # How many points you want in each resampled feature curve

feature_curves_resampled = [resample_curve(curve[0], num_points) for curve in features]
left_curves_resampled = [resample_curve(curve[1], num_points) for curve in features]
right_curves_resampled = [resample_curve(curve[2], num_points) for curve in features]

test_features_curves_resampled = [resample_curve(curve[0], num_points) for curve in test_features]
test_left_curves_resampled = [resample_curve(curve[1], num_points) for curve in test_features]
test_right_curves_resampled = [resample_curve(curve[2], num_points) for curve in test_features]
```

```python
########### CNN MODEL

from tensorflow.keras.models import Model
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np
from tensorflow.keras.callbacks import EarlyStopping

labels = tool.labels.copy()
features = tool.features.copy()
num_points = 100  # How many points you want in each resampled feature curve

feature_curves_resampled = [resample_curve(curve[0], num_points) for curve in features]
left_curves_resampled = [resample_curve(curve[1], num_points) for curve in features]
right_curves_resampled = [resample_curve(curve[2], num_points) for curve in features]

# Reshape your data to be appropriate for a Conv1D layer
# It assumes your *_curves_resampled variables are lists of 1D arrays
feature_curves_array = np.array(feature_curves_resampled)
left_curves_array = np.array(left_curves_resampled)
right_curves_array = np.array(right_curves_resampled)

# Stack the three arrays together along a new axis
# This creates a 3D array of shape (N, 3, 100)
input_array = np.stack((left_curves_array, feature_curves_array, right_curves_array), axis=1)
input_array = input_array / np.max(input_array)

# Convert labels to integers using LabelEncoder
le = LabelEncoder()
labels_encoded = le.fit_transform(labels)

# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(input_array, labels_encoded, test_size=0.2, random_state=42)

# reshape data to fit the CNN input requirements
# Conv2D expects the input to be in the shape (samples, rows, cols, channels)
# since our data is grayscale, we have only 1 channel
X_train = X_train.reshape(X_train.shape[0], 100, 3)
X_test = X_test.reshape(X_test.shape[0], 100, 3)

# define the CNN architecture
model = Sequential()
model.add(Conv1D(32, kernel_size=3, activation='relu', input_shape=(100, 3)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))


# compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# train the model
early_stopping = EarlyStopping(monitor='loss', patience=3)
model.fit(X_train, y_train, epochs=100, batch_size=32, callbacks=[early_stopping])

# evaluate the model
y_pred = model.predict(X_test)
y_pred = np.round(y_pred)  # round the predictions to 0 or 1

# print the classification report
print(classification_report(y_test, y_pred))

from sklearn.metrics import confusion_matrix

# Make the predictions
y_pred_classes = np.argmax(y_pred, axis=1)

# Calculate the confusion matrix
confusion = confusion_matrix(y_test, y_pred_classes)

# Calculate the noise (class 0) and feature (class 1) accuracy
noise_accuracy = confusion[0, 0] / np.sum(confusion[0]) * 100
feature_accuracy = confusion[1, 1] / np.sum(confusion[1]) * 100

print('Noise label accuracy: {:.2f}%'.format(noise_accuracy))
print('Feature label accuracy: {:.2f}%'.format(feature_accuracy))
```
